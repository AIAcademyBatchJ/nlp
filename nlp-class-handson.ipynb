```python
from string import punctuation
```

```python
def clean_txt(sent):
    tokens = word_tokenize(sent.lower())
    stop_updated = stopwords.words('english') + list(punctuation)
    final_words = [x for x in tokens if x not in stop_updated and len(x) > 2]
#     final_words = [str(TextBlob(x).correct()) for x in tokens if x not in stop_updated and len(x) > 2]
    res = ' '.join(final_words)
    return res
```

```python
sent1 = 'This is a very very good, interesting and decent restaurant.'
sent2 = 'This restuarant is very interesting, interesting and awesome!'
```

```python
X = imdb['clean']
y = imdb['sentiment']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2)
```

```python
tfidf_vect = TfidfVectorizer(ngram_range=(1,2))
```

```python
Xtrain_bow = tfidf_vect.fit_transform(Xtrain)
```

```python
Xtest_bow = tfidf_vect.transform(Xtest)
```

```python
Xtrain_bow.toarray()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
```

```python
clf.fit(Xtrain_bow, ytrain)
```

```python
y_pred = clf.predict(Xtest_bow)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(ytest, y_pred)
```

```python
confusion_matrix(ytest, y_pred)
```
